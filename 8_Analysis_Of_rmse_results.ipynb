```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Plaing the regularization parameters and corresponding rmse for the train and test data test into a dataframe
train_rmse_result = []
test_rmse_result = []
for i in range(20):
    file_train = './Files_Folder/rmse/reg/Predicted_train_Result_SVD_mean_' + str(i+1) + '.csv'
    file_test = './Files_Folder/rmse/reg/Predicted_test_Result_SVD_mean_' + str(i+1) + '.csv'
    train_pred = pd.read_csv(file_train)
    test_pred = pd.read_csv(file_test)
    train_rmse_result.append(np.sqrt(train_pred.describe().iloc[1,4]))
    test_rmse_result.append(np.sqrt(test_pred.describe().iloc[1,4]))
reg = [0.01*(j+1) for j in range(20)]
rmse_table = {'reg. param': reg, 'rmse_train': train_rmse_result, 'rmse_test': test_rmse_result}
rmse_table_df = pd.DataFrame(rmse_table)
```

```python
rmse_table_df
```

```python
plt.plot(reg, train_rmse_result, label = 'train', c = 'blue')
plt.plot(reg, test_rmse_result, '--', label = 'test', c = 'red')
plt.xlabel('Regularization Parameters')
plt.ylabel('rmse')
plt.legend()
plt.savefig('rmse_reg.pdf')
```
